```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def My_Edge_Filter(image,sigma):
    image_blur = cv2.GaussianBlur(image,(9,9),sigma)
    sobelx = cv2.Sobel(src=image_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
    sobely = cv2.Sobel(src=image_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
    sobelxy = cv2.Sobel(src=image_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)# Combined X and Y Sobel Edge Detection
    open_kernel = np.ones((3,3))
    sobelxy  = cv2.morphologyEx(sobelxy,cv2.MORPH_OPEN,open_kernel)
    kernel = np.ones((3,3))
    sobelxy = cv2.erode(sobelxy,kernel,iterations = 1)
    return sobelxy
```

```python
image = cv2.imread('Task01.jpg',0)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.xticks([]),plt.yticks([]);
```

```python
new_image = My_Edge_Filter(image,2)
cv2.imshow('Modified Image',new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
